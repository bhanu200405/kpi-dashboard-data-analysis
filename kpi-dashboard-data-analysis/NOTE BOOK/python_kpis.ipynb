```python
import pandas as pd

# Convert Order_Date to datetime
data['Order_Date'] = pd.to_datetime(data['Order_Date'])

# Convert categorical columns
categorical_cols = [
    'Customer_ID', 'Product_ID', 'Category',
    'Sub_Category', 'Brand', 'Region', 'Payment_Mode'
]

for col in categorical_cols:
    data[col] = data[col].astype('category')

# Ensure numeric columns are correct
data['Quantity'] = pd.to_numeric(data['Quantity'], downcast='integer')
data['Unit_Price'] = pd.to_numeric(data['Unit_Price'], downcast='float')
data['Discount'] = pd.to_numeric(data['Discount'], downcast='float')
data['Sales'] = pd.to_numeric(data['Sales'], downcast='float')
data['Profit'] = pd.to_numeric(data['Profit'], downcast='float')
data['Rating'] = pd.to_numeric(data['Rating'], downcast='float')

# Optional: Round rating (if needed)
data['Rating'] = data['Rating'].round(1)

# Check updated info
print(data.info())
```

```python
data['Order_Date'] = pd.to_datetime(data['Order_Date'])
```

```python
##4. Plot monthly sales trend using matplotlib/seaborn
# Create Year-Month column
data['Year_Month'] = data['Order_Date'].dt.to_period('M')

# Group by month and calculate total sales
monthly_sales = data.groupby('Year_Month')['Sales'].sum().reset_index()

# Convert back to timestamp (for plotting)
monthly_sales['Year_Month'] = monthly_sales['Year_Month'].dt.to_timestamp()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(monthly_sales['Year_Month'], monthly_sales['Sales'], marker='o')

plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.grid()

plt.tight_layout()
plt.show()
```

```python
low_rated_products = (
    data.groupby('Product_ID')['Rating']
    .mean()
    .reset_index()
)

low_rated_products = low_rated_products[low_rated_products['Rating'] < 3]

print(low_rated_products)
```

```python

# 6.  profit margins new column
data['Profit_Margin_%'] = np.where(
    data['Sales'] != 0,
    (data['Profit'] / data['Sales']) * 100,
    0
)
```

```python
#7. Filter data orders from the year 2024 
filtered_data = data[
    (data['Order_Date'].dt.year == 2024) &
    (data['Region'] == 'South')
]

print(filtered_data)
```

```python
#8. Top 10 brands by total sales
top_brands = (
    data.groupby('Brand')['Sales']
    .sum()
    .reset_index()
    .sort_values(by='Sales', ascending=False)
    .head(10)
)

print(top_brands)
```

```python
#9. Identify outliers in Unit_Price using IQR method
Q1 = data['Unit_Price'].quantile(0.25)
Q3 = data['Unit_Price'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers_iqr = data[
    (data['Unit_Price'] < lower_bound) | 
    (data['Unit_Price'] > upper_bound)
]

print(outliers_iqr)
```

```python
from scipy.stats import zscore

data['Z_score'] = zscore(data['Unit_Price'])

outliers_z = data[
    (data['Z_score'] > 3) | (data['Z_score'] < -3)
]

print(outliers_z)
```

```python
#10. Summary report by category
summary_report = (
    data.groupby('Category', as_index=False)
    .agg(
        Total_Sales=('Sales', 'sum'),
        Total_Profit=('Profit', 'sum'),
        Total_Quantity=('Quantity', 'sum')
    )
)

print(summary_report)
```
